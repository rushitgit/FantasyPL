```python
import pandas as pd 
import numpy as np
```

```python
xls = pd.ExcelFile("D:/Desktop/FantasyPL/datasets/Player DataSet Final.xlsx")
all_players_data = pd.concat([xls.parse(sheet_name) for sheet_name in xls.sheet_names])
all_players_data = all_players_data.dropna()
all_players_data['xA'] = pd.to_numeric(all_players_data['xA'], errors='coerce')
filtered_data = all_players_data[all_players_data['STATUS'] != 'injured']
filtered_data = filtered_data[filtered_data['OWNERSHIP %'] >= 1]
filtered_data = filtered_data[filtered_data['POINTS'] >= 5]
filtered_data = filtered_data[filtered_data['MINUTES'] >= 50]


filtered_data.to_csv('D:/Desktop/FantasyPL/datasets/PlayerSetCleaned.csv', index=False)
```
